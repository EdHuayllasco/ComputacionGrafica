Ejercicio 4. Guarde una nueva imagen que contenga solo los canales rojo y verde

```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
img_cv = cv2.imread('eiffeltower.jpg') # trabaja con canales BGR
img_rg = img_cv.copy()

img_rg[:,:,0] = 0  # aqui se anula el canal Blue
cv2.imwrite('eiffelrg.png',img_rg)

cv2_imshow(img_rg) # visualizacion del resultado en rojo y verde
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Ejercicio 5. Guarde una nueva imagen que muestre la imagen en escala de grises

```python
img_gris = cv2.imread('eiffeltower.jpg',0) # el segundo par√°metro en cero anula los colores
cv2.imwrite('eiffelgris.png',img_gris)

cv2_imshow(img_gris) # visualizaci√≥n del resultado
cv2.waitKey(0)
cv2.destroyAllWindows()
```
